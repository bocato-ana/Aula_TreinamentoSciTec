2. 📝 Saída de Dados e f-Strings (O Jeito Moderno)
Na Scitec, vamos lidar com muitos dados. O print() com f"" (f-string) permite injetar variáveis dentro de textos de forma limpa.

O f antes das aspas avisa ao Python: "Ei, fique atento às chaves {}!"

```python
nome_projeto = "Dashboard de Vendas"
versao = 1.12

# Usando f-strings com emojis para relatórios visuais
print(f"🚀 Iniciando projeto: {nome_projeto}")
print(f"📌 Versão atual: {versao:.1f}") # o :.1f força uma casa decimal
```
