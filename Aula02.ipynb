📚 7. Coleções de Dados: Listas vs. Tuplas
Na Scitec, raramente trabalhamos com um dado sozinho. Usamos coleções para agrupar informações.

📝 Listas ([]) - As Flexíveis
As listas são mutáveis. Você pode adicionar, remover e alterar itens o tempo todo. Use para dados que mudam (ex: lista de tarefas, sensores ativos).

```python
# Criando uma lista de membros da Scitec
equipe_scitec = ["Ana", "João", "Maria"]
equipe_scitec.append("Carlos") # Adicionando um novo membro
equipe_scitec[1] = "João Silva" # Alterando um nome

print(f"👥 Equipe atualizada: {equipe_scitec}")
```

🔒 Tuplas (()) - As Imutáveis
As tuplas são imutáveis. Uma vez criadas, você não pode alterar nada. Elas são mais rápidas e seguras para dados que nunca devem mudar (ex: coordenadas GPS, meses do ano, configurações fixas).

```python
# Criando uma tupla com as siglas dos departamentos
departamentos = ("Dados", "Projetos", "RH", "Marketing")
t: tuple[int,...] = (1, 2, 3, 4)

# Tentativa de erro (Descomente para ver o Python protegendo o dado):
# departamentos[0] = "TI" 

print(f"🏢 Departamentos fixos: {departamentos}")
print(t)
```

🚦 8. Estruturas de Controle: O Cérebro do Código (if/else e while)
Para que o seu código tome decisões sozinho, usamos a lógica condicional e a repetição.

🛣️ O Caminho do if e else
Pense nisso como uma bifurcação na estrada.

```python
# Verificador de Meta de Dados
processados: int = 450
meta: int = 500

if processados >= meta:
    print("✅ Meta batida! Parabéns, Scitec!")
elif processados >= (meta * 0.8):
    print("⚠️ Quase lá! Faltam poucos registros.")
else:
    print("❌ Alerta: Produção abaixo do esperado.")
```

🔁 O Ciclo do while
Diferente do for, o while continua rodando enquanto uma condição for verdadeira. É perfeito para criar menus ou esperar um sinal do usuário.

```python
# Exemplo de loop de segurança
tentativas: int = 0
limite: int = 3

while tentativas < limite:
    tentativas += 1
    print(f"🔄 Tentativa {tentativas} de conexão com o banco de dados...")

print("📡 Conexão estabelecida!")
```

🏆 Desafio Final: O Sistema da Scitec
Combine tudo o que aprendemos hoje!

Instruções:

Crie uma lista chamada valores_brutos com os números: [10, 50, 150, 20, 300].

Crie uma função com Type Hints que percorra essa lista.

Se o valor for maior que 100, coloque-o em uma nova lista chamada alertas.

No final, use um if para printar: "Sistema Seguro" (se não houver alertas) ou "⚠️ Revisar valores: [lista]" (se houver).

```python
# 1. Definição dos dados iniciais (Casting e Type Hinting)
valores_brutos: list = [10, 50, 150, 20, 300]

# 2. Criando a função com Type Hints (Contrato de Confiança)
def analisar_integridade(dados: list) -> list:
    """
    Recebe uma lista de valores e retorna apenas aqueles 
    que ultrapassam o limite crítico de 100.
    """
    # Usando List Comprehension (Nível Intermediário!)
    alertas = [valor for valor in dados if valor > 100]
    return alertas

# 3. Executando a lógica
lista_de_alertas = analisar_integridade(valores_brutos)

# 4. Estrutura de Decisão (if/else) para o relatório final
if not lista_de_alertas: # Verifica se a lista está vazia
    print("✅ Sistema Seguro: Nenhum valor crítico detectado.")
else:
    print(f"⚠️ Atenção Scitec! Revisar os seguintes valores: {lista_de_alertas}")
```
